```python
import pandas as pd
```

```python
print("\n========== LOADING DATASET ==========\n")

# Store the file path in a variable so it can be changed easily later.
file_path = "../data/raw/Credit card transactions - India - Simple.csv"

df = pd.read_csv(file_path)

print("Dataset loaded successfully.")
print()
```

```python
# Shape returns number of rows and columns.
rows, cols = df.shape

print(f"Number of rows       : {rows}")
print(f"Number of columns    : {cols}")
print()
```

```python
# Loop through all columns and print one by one.
for column in df.columns:
    print(column)

print()
```

```python
print(df.dtypes)

print()
```

```python
# Head shows the first few rows.
print(df.head())

print()
```

```python
# Tail shows the last few rows.
print(df.tail())

print()
```

```python
# isnull() checks for missing values.
# sum() counts them column by column.
print(df.isnull().sum())

print()
```

```python
# describe() generates statistical summaries for numeric columns.
print(df.describe())

print()
```

```python
columns_to_check = [
    "City",
    "Card Type",
    "Exp Type",
    "Gender"
]

# Loop through each column and print value counts
for column in columns_to_check:

    # Count how many times each category appears
    print(df[column].value_counts())

print()
```

```python
# info() prints memory usage, columns and non-null counts.
print(df.info())

print()
```
