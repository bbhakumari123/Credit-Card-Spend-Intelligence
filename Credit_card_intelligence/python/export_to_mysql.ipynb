```python
import pandas as pd

# Import SQLAlchemy engine
from sqlalchemy import create_engine
```

```python
pythonengine = create_engine(
    "mysql+mysqlconnector://root:your_password@localhost/creditcard_db"
)

print("Connection string created.\n")
```

```python
clean_df = pd.read_csv(
    "../data/cleaned/credit_card_cleaned.csv"
)

# Summary tables
city_summary = pd.read_csv(
    "../outputs/csv_exports/city_summary.csv"
)

category_summary = pd.read_csv(
    "../outputs/csv_exports/category_summary.csv"
)

gender_summary = pd.read_csv(
    "../outputs/csv_exports/gender_summary.csv"
)

monthly_summary = pd.read_csv(
    "../outputs/csv_exports/monthly_summary.csv"
)

stat_results = pd.read_csv(
    "../outputs/statistical_results/statistical_test_results.csv"
)
print("All files loaded successfully.\n")
```

```python
clean_df.to_sql(
    name="credit_card_master",
    con=pythonengine,
    if_exists="replace",
    index=False
)

print("credit_card_master exported.\n")
```

```python

city_summary.to_sql(
    name="city_summary",
    con=pythonengine,
    if_exists="replace",
    index=False
)

print("city_summary exported.\n")
```

```python
category_summary.to_sql(
    name="category_summary",
    con=pythonengine,
    if_exists="replace",
    index=False
)

print("category_summary exported.\n")
```

```python
gender_summary.to_sql(
    name="gender_summary",
    con=pythonengine,
    if_exists="replace",
    index=False
)

print("gender_summary exported.\n")
```

```python
monthly_summary.to_sql(
    name="monthly_summary",
    con=pythonengine,
    if_exists="replace",
    index=False
)

print("monthly_summary exported.\n")
```

```python
stat_results.to_sql(
    name="statistical_results",
    con=pythonengine,
    if_exists="replace",
    index=False
)

print("statistical_results exported.\n")
```
