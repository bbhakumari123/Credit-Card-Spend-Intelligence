```python
import pandas as pd
```

```python
file_path = "../data/cleaned/credit_card_cleaned.csv"

df = pd.read_csv(file_path)

print("Cleaned dataset loaded successfully.")
print()
```

```python
city_summary = (
    df.groupby("City")["Amount"]
    .sum()
    .reset_index()
    .sort_values(by="Amount", ascending=False)
)

print(city_summary)

print()
```

```python
print("========== TOTAL SPEND BY EXPENSE TYPE ==========\n")

category_summary = (
    df.groupby("Exp Type")["Amount"]
    .sum()
    .reset_index()
    .sort_values(by="Amount", ascending=False)
)

print(category_summary)

print()
```

```python
card_summary = (
    df.groupby("Card Type")["Amount"]
    .sum()
    .reset_index()
    .sort_values(by="Amount", ascending=False)
)

print(card_summary)

print()
```

```python
gender_summary = (
    df.groupby("Gender")["Amount"]
    .sum()
    .reset_index()
    .sort_values(by="Amount", ascending=False)
)

print(gender_summary)

print()
```

```python
monthly_summary = (
    df.groupby(
        ["Year", "Month_Number", "Month"]
    )["Amount"]
    .sum()
    .reset_index()
    .sort_values(
        by=["Year", "Month_Number"]
    )
)

print(monthly_summary)

print()
```

```python
city_category_summary = (
    df.groupby(["City", "Exp Type"])["Amount"]
    .sum()
    .reset_index()
    .sort_values(by="Amount", ascending=False)
)

print(city_category_summary.head(20))

print()
```

```python
city_summary.to_csv(
    "../outputs/csv_exports/city_summary.csv",
    index=False
)
```

```python
category_summary.to_csv(
    "../outputs/csv_exports/category_summary.csv",
    index=False
)

card_summary.to_csv(
    "../outputs/csv_exports/card_summary.csv",
    index=False
)

gender_summary.to_csv(
    "../outputs/csv_exports/gender_summary.csv",
    index=False
)

monthly_summary.to_csv(
    "../outputs/csv_exports/monthly_summary.csv",
    index=False
)

city_category_summary.to_csv(
    "../outputs/csv_exports/city_category_summary.csv",
    index=False
)
```
