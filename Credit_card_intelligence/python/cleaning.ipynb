```python
import pandas as pd
```

```python
file_path = "../data/raw/Credit card transactions - India - Simple.csv"

# Read the CSV into a pandas DataFrame
df = pd.read_csv(file_path)

print("Dataset loaded successfully.")
print()
```

```python
print(df.columns.tolist())

print()
```

```python
df.columns = df.columns.str.strip()

print("Column names standardized.")
print()
```

```python
df["Date"] = pd.to_datetime(df["Date"])

print("Date column converted successfully.")
print(df["Date"].head())

print()
```

```python
df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")

print("Amount converted to numeric datatype.")
print(df["Amount"].dtype)

print()
```

```python
# Extract year from Date
df["Year"] = df["Date"].dt.year

print("Year column created.")
```

```python
# Extract month name from Date
df["Month"] = df["Date"].dt.month_name()

print("Month column created.")
```

```python
# Month number helps sort months correctly
df["Month_Number"] = df["Date"].dt.month

print("Month_Number column created.")
```

```python
df["Quarter"] = df["Date"].dt.quarter

print("Quarter column created.")
print()
```

```python
# Remove extra spaces and use title case
# Example: mumbai -> Mumbai
df["City"] = df["City"].str.strip().str.title()

print("City names standardized.")

print()
```

```python
print(df.info())

print()
```

```python
# Output file path
output_path = "../data/cleaned/credit_card_cleaned.csv"

# Save cleaned data without index column
df.to_csv(output_path, index=False)

print(f"Cleaned file saved successfully at:")
print(output_path)

print()
```
