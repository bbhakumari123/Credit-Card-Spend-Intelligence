```python
import pandas as pd
from scipy.stats import ttest_ind
from scipy.stats import chi2_contingency
from scipy.stats import f_oneway
from scipy.stats import pearsonr
```

```python
file_path = "../data/cleaned/credit_card_cleaned.csv"

df = pd.read_csv(file_path)

print("Dataset loaded successfully.\n")

results = []
```

```python
male_spend = df[df["Gender"] == "M"]["Amount"]

female_spend = df[df["Gender"] == "F"]["Amount"]

t_stat, p_value = ttest_ind(
    male_spend,
    female_spend,
    equal_var=False
)

print("T Statistic :", round(t_stat, 4))
print("P Value     :", round(p_value, 6))


if p_value < 0.05:
    interpretation = (
        "There is a statistically significant difference "
        "between male and female spending."
    )
else:
    interpretation = (
        "No statistically significant difference "
        "between male and female spending."
    )

print("Interpretation:")
print(interpretation)
print()


results.append({
    "Test": "T-Test",
    "Statistic": t_stat,
    "P_Value": p_value,
    "Interpretation": interpretation
})
```

```python
contingency_table = pd.crosstab(
    df["City"],
    df["Exp Type"]
)

chi2, p_value, dof, expected = chi2_contingency(
    contingency_table
)

print("Chi-Square Statistic :", round(chi2, 4))
print("P Value              :", round(p_value, 6))


if p_value < 0.05:
    interpretation = (
        "Expense category depends on city."
    )
else:
    interpretation = (
        "Expense category is independent of city."
    )

print("Interpretation:")
print(interpretation)
print()


results.append({
    "Test": "Chi-Square",
    "Statistic": chi2,
    "P_Value": p_value,
    "Interpretation": interpretation
})
```

```python
q1 = df[df["Quarter"] == 1]["Amount"]
q2 = df[df["Quarter"] == 2]["Amount"]
q3 = df[df["Quarter"] == 3]["Amount"]
q4 = df[df["Quarter"] == 4]["Amount"]

f_stat, p_value = f_oneway(
    q1,
    q2,
    q3,
    q4
)

print("F Statistic :", round(f_stat, 4))
print("P Value     :", round(p_value, 6))


if p_value < 0.05:
    interpretation = (
        "Average spending differs significantly across quarters."
    )
else:
    interpretation = (
        "No significant spending difference across quarters."
    )

print("Interpretation:")
print(interpretation)
print()


results.append({
    "Test": "ANOVA",
    "Statistic": f_stat,
    "P_Value": p_value,
    "Interpretation": interpretation
})
```

```python
correlation, p_value = pearsonr(
    df["Amount"],
    df["Month_Number"]
)

print("Correlation :", round(correlation, 4))
print("P Value     :", round(p_value, 6))


if abs(correlation) < 0.3:
    interpretation = (
        "Weak relationship between transaction amount and month."
    )

elif abs(correlation) < 0.7:
    interpretation = (
        "Moderate relationship between transaction amount and month."
    )

else:
    interpretation = (
        "Strong relationship between transaction amount and month."
    )


print("Interpretation:")
print(interpretation)
print()


results.append({
    "Test": "Pearson Correlation",
    "Statistic": correlation,
    "P_Value": p_value,
    "Interpretation": interpretation
})
```

```python
results_df = pd.DataFrame(results)

output_path = (
    "../outputs/statistical_results/"
    "statistical_test_results.csv"
)

results_df.to_csv(
    output_path,
    index=False
)

print("Statistical results exported successfully.")
print(output_path)

print()
```
